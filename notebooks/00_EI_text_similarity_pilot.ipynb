# Get text similarity

```python
# apply to all LMs
from scipy import stats
def cosine_similarity(a, b):
  cos_sim = round(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)), 3)
  #print('cos_sim', cos_sim)
  return cos_sim

def spearman_similarity(a, b):
  sp = round(stats.spearmanr(a, b)[0], 3)
  #print('spearman_sim', sp)
  return sp
```
